```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# generate random numbers from a skewed normal distribution

x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color = "green")
```

## range

```python
max_value = np.max(x)
min_value = np.min(x)
```

```python
print(f"Maximum Value: {max_value}")
print(f"Maximum Value: {min_value}")
```

```python
range = max_value - min_value
print(f"range: {range}")
```

## Inter Quartile Range

```python
q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)
```

```python
iqr = q3 - q1
print(f"Interquartile range: {iqr}")
```

## Mean absolute deviation

```python
mad = np.mean(np.abs(x - np.mean(x)))
print("MAD: ", mad)
```

```python
plt.hist(x, color = "green")
plt.axvline(x = mad, color="red")
plt.axvline(x = np.mean(x), color='yellow')
```

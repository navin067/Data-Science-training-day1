```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# generate random numbers from a skewed normal distribution

x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color = "green")
```

## Variance

```python
variance = np.var(x)
print("Variance: ", variance)
```

```python
plt.hist(x, color = "green")
plt.axvline(x = x.mean(), color = "yellow")
plt.axvline(x = variance, color = "blue")
```

## Standard deviation

```python
std = np.std(x)
print("Standard deviation: ",std)
```

```python
plt.hist(x, color="green")
plt.axvline(x = std, color = "blue")
plt.axvline(x = x.mean(), color = "yellow")
```
